# user input

```python
userInput = [
            {'ProductName':'سیب زمینی متوسط دست چین کیلویی', 'Count':18},
            {'ProductName':'لیمو ترش سنگی دستچین کیلویی', 'Count':6},
            {'ProductName':'کاهو رسمی دستچین کیلویی', 'Count':40},
            {'ProductName':"کلم بروکلی دستچین کیلویی", 'Count':5},
            {'ProductName':'قارچ دکمه ای سفید دستچین و تازه کیلویی', 'Count':10}
         ] 
inputorder = pd.DataFrame(userInput)
inputorder
```

```python
#Filtering out the product by ProductName
inputId = product_df[product_df['ProductName'].isin(inputorder['ProductName'].tolist())]
#Then merging it so we can get the  ProductId. It's implicitly merging it by ProductName.
inputproduct = pd.merge(inputId, inputorder)

#Final input dataframe
#If a product you added in above isn't here, then it might not be in the original 
#dataframe or it might spelled differently, please check capitalisation.
inputproduct
```

## grouping users

```python
#Filtering out users that have watched products that the input has watched and storing it
userSubset = orders_df[orders_df['ProductId'].isin(inputproduct['ProductId'].tolist())]
userSubset.head()
```

```python
#Groupby creates several sub dataframes where they all have the same value in the column specified as the parameter
userSubsetGroup = userSubset.groupby(['BuyerId'])
```

### example

### sort the groups

```python
#Sorting it so users with product most in common with the input will have priority
userSubsetGroup = sorted(userSubsetGroup,  key=lambda x: len(x[1]), reverse=True)
```

```python
userSubsetGroup[0:3]
```

```python
userSubsetGroup = userSubsetGroup[0:50]
```

```python
#Store the Pearson Correlation in a dictionary, where the key is the user Id and the value is the coefficient
pearsonCorrelationDict = {}

#For every user group in our subset
for name, group in userSubsetGroup:
    #Let's start by sorting the input and current user group so the values aren't mixed up later on
    group = group.sort_values(by='ProductId')
    inputproduct = inputproduct.sort_values(by='ProductId')
    #Get the N for the formula
    nRatings = len(group)
    #Get the review scores for the movies that they both have in common
    temp_df = inputproduct[inputproduct['ProductId'].isin(group['ProductId'].tolist())]
    #And then store them in a temporary buffer variable in a list format to facilitate future calculations
    tempRatingList = temp_df['Count'].tolist()
    #Let's also put the current user group reviews in a list format
    tempGroupList = group['Count'].tolist()
    #Now let's calculate the pearson correlation between two users, so called, x and y
    Sxx = sum([i**2 for i in tempRatingList]) - pow(sum(tempRatingList),2)/float(nRatings)
    Syy = sum([i**2 for i in tempGroupList]) - pow(sum(tempGroupList),2)/float(nRatings)
    Sxy = sum( i*j for i, j in zip(tempRatingList, tempGroupList)) - sum(tempRatingList)*sum(tempGroupList)/float(nRatings)
    
    #If the denominator is different than zero, then divide, else, 0 correlation.
    if Sxx != 0 and Syy != 0:
        pearsonCorrelationDict[name] = Sxy/sqrt(Sxx*Syy)
    else:
        pearsonCorrelationDict[name] = 0
```
